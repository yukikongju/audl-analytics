```python
name = soup.find('div', class_='audl-player-display-name').text.strip()
team_position = soup.find('div', class_='audl-player-current-team-position').text.strip()
jersey_number = soup.find('div', class_='audl-player-jersey-number').text.strip()
# team = team_position.split('/')[0]
# position = team_position.split('/')[1]

# Extract personal stats
stats = {}
for item in soup.find_all('div', class_='audl-personal-stats-item'):
    label = item.find('span', class_='audl-personal-stats-label').text.strip()
    value = item.find('span', class_='audl-personal-stats-value').text.strip()
    stats[label] = value

# Extract player bio text and hometown
bio_paragraphs = soup.select('.audl-player-bio-text p')
bio_text = ' '.join([p.text for p in bio_paragraphs if 'Hometown:' not in p.text])
hometown = ''
for p in bio_paragraphs:
    if 'Hometown:' in p.text:
        hometown = p.text.split('Hometown:')[-1].strip()

# Combine all into a dictionary
player_data = {
    'Name': name,
    'Team': team,
    'Team / Position': team_position, 
    # 'Position': position,
    'Jersey Number': jersey_number,
    'Height': stats.get('HEIGHT'),
    'Weight': stats.get('WEIGHT'),
    'DOB': stats.get('AGE/DOB'),
    'College': stats.get('COLLEGE'),
    'Hometown': hometown,
    'Bio': bio_text
}

# Convert to DataFrame
df = pd.DataFrame([player_data])
# print(df)
```

### Upsert

```python
from sqlalchemy import Table, Column, Integer, String, Boolean, MetaData
from sqlalchemy.dialects.postgresql import insert
from sqlalchemy.sql import literal_column

metadata = MetaData()

distributors = Table(
    "distributors", metadata,
    Column("did", Integer, primary_key=True),
    Column("dname", String),
    Column("is_active", Boolean, default=True)
)

row = {"did": 10, "dname": "Conrad International"}

# Build base insert
stmt = insert(distributors).values(**row)

# Dynamically build the SET clause (exclude primary key)
update_dict = {
    k: getattr(stmt.excluded, k)
    for k in row.keys()
    if k != "did"  # assuming 'did' is the conflict target / PK
}

# Create upsert statement with WHERE clause on conflict
upsert_stmt = stmt.on_conflict_do_update(
    index_elements=["did"],
    index_where=literal_column("is_active"),  # Partial index condition
    set_=update_dict
)

# Execute
with engine.begin() as conn:
    conn.execute(upsert_stmt)
```

### Playwright

```python
from playwright.sync_api import sync_playwright

def scrape_players():
    with sync_playwright() as p:
        # Launch a browser instance (headless mode for faster scraping)
        browser = p.chromium.launch(headless=True)
        page = browser.new_page()
        
        # Navigate to the target URL
        page.goto("https://www.watchufa.com/league/players")
        
        # Wait for the page to load fully (adjust selector if needed)
        page.wait_for_selector("table")  # Assuming data is in a table
        
        # Extract player data
        rows = page.query_selector_all("table tr")
        players = []
        
        for row in rows[1:]:  # Skip header row
            columns = row.query_selector_all("td")
            player_name = columns[0].inner_text()
            recent_team = columns[1].inner_text()
            players.append({"Player": player_name, "Most Recent Team": recent_team})
        
        browser.close()
        return players

# Execute the function and print results
players_data = scrape_players()
for player in players_data:
    print(player)
```

```python
import asyncio
from playwright.async_api import async_playwright

async def scrape_players():
    async with async_playwright() as p:
        # Launch a browser instance (headless mode for faster scraping)
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()
        
        # Navigate to the target URL
        await page.goto("https://www.watchufa.com/league/players")
        
        # Wait for the page to load fully (adjust selector if needed)
        await page.wait_for_selector("table")  # Assuming data is in a table
        
        # Extract player data
        rows = await page.query_selector_all("table tr")
        players = []
        
        for row in rows[1:]:  # Skip header row
            columns = await row.query_selector_all("td")
            player_name = await columns[0].inner_text()
            recent_team = await columns[1].inner_text()
            players.append({"Player": player_name, "Most Recent Team": recent_team})
        
        await browser.close()
        return players

# Run the async function
async def main():
    players_data = await scrape_players()
    for player in players_data:
        print(player)

# Start asyncio event loop
asyncio.run(main())
```
